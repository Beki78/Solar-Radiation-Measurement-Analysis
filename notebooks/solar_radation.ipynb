## Visualize solar radiation and temperature over time:

```python
import matplotlib.pyplot as plt

# Convert Timestamp to datetime
benin['Timestamp'] = pd.to_datetime(benin['Timestamp'])

# Plot GHI, DNI, DHI over time
plt.figure(figsize=(14, 7))
plt.plot(benin['Timestamp'], benin['GHI'], label='GHI (W/mÂ²)')
plt.plot(benin['Timestamp'], benin['DNI'], label='DNI (W/mÂ²)')
plt.plot(benin['Timestamp'], benin['DHI'], label='DHI (W/mÂ²)')
plt.xlabel('Date')
plt.ylabel('Irradiance (W/mÂ²)')
plt.title('Solar Radiation Over Time')
plt.legend()
plt.show()
```
