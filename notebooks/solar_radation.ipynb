# Wind Analysis

```python
def plot_wind_rose(benin):
    wind_speed = benin['WS']
    wind_direction = benin['WD']
    
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, projection='polar')
    ax.hist(np.radians(wind_direction), bins=30, weights=wind_speed, alpha=0.5)
    ax.set_title('Wind Rose')
    plt.show()

plot_wind_rose(benin)
```

```python
def plot_wind_rose(serraleone):
    wind_speed = serraleone['WS']
    wind_direction = serraleone['WD']
    
    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, projection='polar')
    ax.hist(np.radians(wind_direction), bins=30, weights=wind_speed, alpha=0.5)
    ax.set_title('Wind Rose')
    plt.show()

plot_wind_rose(serraleone)
```
